```python
# Utiltiy Functions

def get_crops(frame: np.ndarray, detections: sv.Detections) -> List[np.ndarray]:
    """
    Extract crops from the frame based on detected bounding boxes.

    Args:
        frame (np.ndarray): The frame from which to extract crops.
        detections (sv.Detections): Detected objects with bounding boxes.

    Returns:
        List[np.ndarray]: List of cropped images.
    """
    return [sv.crop_image(frame, xyxy) for xyxy in detections.xyxy]


def resolve_goalkeepers_team_id(
    players: sv.Detections,
    players_team_id: np.array,
    goalkeepers: sv.Detections
) -> np.ndarray:
    """
    Resolve the team IDs for detected goalkeepers based on the proximity to team
    centroids.

    Args:
        players (sv.Detections): Detections of all players.
        players_team_id (np.array): Array containing team IDs of detected players.
        goalkeepers (sv.Detections): Detections of goalkeepers.

    Returns:
        np.ndarray: Array containing team IDs for the detected goalkeepers.

    This function calculates the centroids of the two teams based on the positions of
    the players. Then, it assigns each goalkeeper to the nearest team's centroid by
    calculating the distance between each goalkeeper and the centroids of the two teams.
    """
    goalkeepers_xy = goalkeepers.get_anchors_coordinates(sv.Position.BOTTOM_CENTER)
    players_xy = players.get_anchors_coordinates(sv.Position.BOTTOM_CENTER)
    team_0_centroid = players_xy[players_team_id == 0].mean(axis=0)
    team_1_centroid = players_xy[players_team_id == 1].mean(axis=0)
    goalkeepers_team_id = []
    for goalkeeper_xy in goalkeepers_xy:
        dist_0 = np.linalg.norm(goalkeeper_xy - team_0_centroid)
        dist_1 = np.linalg.norm(goalkeeper_xy - team_1_centroid)
        goalkeepers_team_id.append(0 if dist_0 < dist_1 else 1)
    return np.array(goalkeepers_team_id)


```

```python
def render_radar(
    detections: sv.Detections,
    keypoints: sv.KeyPoints,
    color_lookup: np.ndarray,
    ball_positions: np.ndarray
) -> np.ndarray:
    mask = (keypoints.xy[0][:, 0] > 1) & (keypoints.xy[0][:, 1] > 1)
    transformer = ViewTransformer(
        source=keypoints.xy[0][mask].astype(np.float32),
        target=np.array(CONFIG.vertices)[mask].astype(np.float32)
    )
    xy = detections.get_anchors_coordinates(anchor=sv.Position.BOTTOM_CENTER)
    transformed_xy = transformer.transform_points(points=xy)

    transformed_ball_positions = transformer.transform_points(points=ball_positions)

    radar = draw_pitch(config=CONFIG)
    radar = draw_points_on_pitch(
        config=CONFIG, xy=transformed_xy[color_lookup == 0],
        face_color=sv.Color.from_hex(COLORS[0]), radius=20, pitch=radar)
    radar = draw_points_on_pitch(
        config=CONFIG, xy=transformed_xy[color_lookup == 1],
        face_color=sv.Color.from_hex(COLORS[1]), radius=20, pitch=radar)
    radar = draw_points_on_pitch(
        config=CONFIG, xy=transformed_xy[color_lookup == 2],
        face_color=sv.Color.from_hex(COLORS[2]), radius=20, pitch=radar)
    radar = draw_points_on_pitch(
        config=CONFIG, xy=transformed_xy[color_lookup == 3],
        face_color=sv.Color.from_hex(COLORS[3]), radius=20, pitch=radar)
    radar = draw_points_on_pitch(
        config=CONFIG, xy=transformed_ball_positions,
        face_color=sv.Color.from_hex(BALL_COLOR), radius=10, pitch=radar)
    return radar
```

```python
from typing import List , Dict , Any

def run_radar_with_json(source_video_path: str, device: str) -> Iterator[Dict[str, Any]]:
    # Initialization of Models
    player_detection_model = YOLO(PLAYER_DETECTION_MODEL_PATH).to(device=device)
    pitch_detection_model = YOLO(PITCH_DETECTION_MODEL_PATH).to(device=device)
    ball_detection_model = YOLO(BALL_DETECTION_MODEL_PATH).to(device=device)

    # Reading the Video Frames
    frame_generator = sv.get_video_frames_generator(
        source_path=source_video_path, stride=STRIDE
    )

    # Collecting Crops for Training
    crops = []
    for frame in tqdm(frame_generator, desc='collecting crops'):
        result = player_detection_model(frame, imgsz=1280, verbose=False)[0]
        detections = sv.Detections.from_ultralytics(result)
        crops += get_crops(frame, detections[detections.class_id == PLAYER_CLASS_ID])

    # Training the Team Classifier
    team_classifier = TeamClassifier(device=device)
    team_classifier.fit(crops)

    # Set Up for Frame Analysis
    frame_generator = sv.get_video_frames_generator(
        source_path=source_video_path, stride=30
    )
    # tracker = sv.ByteTrack(minimum_consecutive_frames=5)

    # Ball Tracking Setup
    ball_tracker = BallTracker(buffer_size=20)
    ball_annotator = BallAnnotator(radius=6, buffer_size=10)

    # Define Ball Detection Slicer
    def ball_callback(image_slice: np.ndarray) -> sv.Detections:
        result = ball_detection_model(image_slice, imgsz=640, verbose=False)[0]
        return sv.Detections.from_ultralytics(result)

    slicer = sv.InferenceSlicer(
        callback=ball_callback,
        slice_wh=(640, 640),
        # overlap_wh=(0.2, 0.2),  # Adjust overlap as needed
    )

    # Process video frames
    for frame_index, frame in enumerate(frame_generator):
        try:
            timestamp = f"{frame_index / 30:.2f}"  # Assuming 30 FPS video

            # Pitch detection and keypoints
            result = pitch_detection_model(frame, verbose=False)[0]
            keypoints = sv.KeyPoints.from_ultralytics(result)

            # Player detections
            result = player_detection_model(frame, imgsz=1280, verbose=False)[0]
            detections = sv.Detections.from_ultralytics(result)
            players = detections[detections.class_id == PLAYER_CLASS_ID]
            players_team_id = team_classifier.predict(get_crops(frame, players))

            # Goalkeeper and referee detections
            goalkeepers = detections[detections.class_id == GOALKEEPER_CLASS_ID]
            referees = detections[detections.class_id == REFEREE_CLASS_ID]

            # Ball detections
            ball_detections = slicer(frame).with_nms(threshold=0.5)
            ball_positions = ball_detections.get_anchors_coordinates(anchor=sv.Position.BOTTOM_CENTER)

            # Prepare JSON data
            frame_data = {
                "frame_index": frame_index,
                "timestamp": timestamp,
                "detections": {
                    "players": [
                        {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2), "team_id": int(team_id)}
                        for pos, conf, team_id in zip(players.xyxy, players.confidence, players_team_id)
                    ],
                    "goalkeepers": [
                        {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2), "team_id": int(team_id)}
                        for pos, conf, team_id in zip(goalkeepers.xyxy, goalkeepers.confidence, players_team_id)
                    ],
                    "referees": [
                        {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2)}
                        for pos, conf in zip(referees.xyxy, referees.confidence)
                    ],
                    "ball": [
                        {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2)}
                        for pos, conf in zip(ball_positions, ball_detections.confidence)
                    ]
                },
                "keypoints": keypoints.to_dict() if hasattr(keypoints, "to_dict") else {},
            }

            # annotated_frame = sv.draw_image(frame, radar, opacity=1)
            yield {"json": frame_data}

        except Exception as e:
            print(f"Error processing frame: {e}")
            continue

```

```python
import cv2
import numpy as np
import json
import uuid
from typing import List, Dict, Any, Iterator, Callable
from itertools import count
from tqdm import tqdm
import supervision as sv

# Initialize ByteTrack tracker
tracker = sv.ByteTrack()

# Helper function to generate unique player IDs
def generate_player_id() -> str:
    return str(uuid.uuid4())

# Convert NumPy types to Python types for JSON serialization
def convert_np_types(obj: Any) -> Any:
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, np.generic):
        return obj.item()
    elif isinstance(obj, dict):
        return {key: convert_np_types(value) for key, value in obj.items()}
    elif isinstance(obj, list):
        return [convert_np_types(item) for item in obj]
    else:
        return obj

# Callback for player detection
def detect_players(frame, model, team_classifier) -> Dict[str, Any]:
    result = model(frame, imgsz=1280, verbose=False)[0]
    detections = sv.Detections.from_ultralytics(result)
    players = detections[detections.class_id == PLAYER_CLASS_ID]
    players_team_id = team_classifier.predict(get_crops(frame, players))
    return {"players": players, "players_team_id": players_team_id}

# Callback for pitch detection
def detect_pitch(frame: np.ndarray, model) -> sv.KeyPoints:
    result = model(frame, verbose=False)[0]
    return sv.KeyPoints.from_ultralytics(result)

# Callback for ball detection
def detect_ball(frame: np.ndarray, slicer: sv.InferenceSlicer) -> Dict[str, Any]:
    ball_detections = slicer(frame).with_nms(threshold=0.5)
    ball_positions = ball_detections.get_anchors_coordinates(anchor=sv.Position.BOTTOM_CENTER)
    return {"ball_detections": ball_detections, "ball_positions": ball_positions}

# Main processing pipeline
def process_video_with_callbacks(
    source_video_path: str,
    device: str,
    player_detection_callback: Callable,
    pitch_detection_callback: Callable,
    ball_detection_callback: Callable
) -> Iterator[Dict[str, Any]]:
    # Initialize models
    player_detection_model = YOLO(PLAYER_DETECTION_MODEL_PATH).to(device=device)
    pitch_detection_model = YOLO(PITCH_DETECTION_MODEL_PATH).to(device=device)
    ball_detection_model = YOLO(BALL_DETECTION_MODEL_PATH).to(device=device)

    # Train the team classifier
    frame_generator = sv.get_video_frames_generator(source_path=source_video_path, stride=STRIDE)
    crops = []
    for frame in tqdm(frame_generator, desc="Collecting crops"):
        result = player_detection_model(frame, imgsz=1280, verbose=False)[0]
        detections = sv.Detections.from_ultralytics(result)
        crops += get_crops(frame, detections[detections.class_id == PLAYER_CLASS_ID])
    team_classifier = TeamClassifier(device=device)
    team_classifier.fit(crops)

    # Setup ball detection slicer
    def ball_callback(image_slice: np.ndarray) -> sv.Detections:
        result = ball_detection_model(image_slice, imgsz=640, verbose=False)[0]
        return sv.Detections.from_ultralytics(result)

    slicer = sv.InferenceSlicer(callback=ball_callback, slice_wh=(640, 640))

    # Process frames
    frame_generator = sv.get_video_frames_generator(source_path=source_video_path, stride=30)
    for frame_index, frame in enumerate(frame_generator):
        timestamp = f"{frame_index / 30:.2f}"  # Assuming 30 FPS video

        # Run callbacks
        pitch_data = pitch_detection_callback(frame, pitch_detection_model)
        player_data = player_detection_callback(frame, player_detection_model, team_classifier)
        ball_data = ball_detection_callback(frame, slicer)

        # Generate JSON data
        frame_data = {
            "frame_index": frame_index,
            "timestamp": timestamp,
            "detections": {
                "players": [
                    {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2), "team_id": int(team_id)}
                    for pos, conf, team_id in zip(
                        player_data["players"].xyxy,
                        player_data["players"].confidence,
                        player_data["players_team_id"]
                    )
                ],
                "ball": [
                    {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2)}
                    for pos, conf in zip(
                        ball_data["ball_positions"],
                        ball_data["ball_detections"].confidence
                    )
                ]
            },
            "keypoints": pitch_data.to_dict() if hasattr(pitch_data, "to_dict") else {},
        }

        yield {"json": frame_data}

# Process video and save to JSON
all_frame_data = []
for result in process_video_with_callbacks(
    source_video_path,
    device,
    detect_players,
    detect_pitch,
    detect_ball
):
    frame_data = result["json"]
    print(frame_data)
    all_frame_data.append(convert_np_types(frame_data))

json_output_path = "output01.json"
with open(json_output_path, "w") as json_file:
    json.dump(all_frame_data, json_file, indent=4)


```

```python
import cv2
import numpy as np
import json
import uuid
from typing import List, Dict, Any, Iterator, Callable
from tqdm import tqdm
import supervision as sv
from itertools import count

# Initialize ByteTrack tracker
tracker = sv.ByteTrack()

# Helper function to generate unique player IDs
def generate_player_id() -> str:
    return str(uuid.uuid4())

# Convert NumPy types to Python types for JSON serialization
def convert_np_types(obj: Any) -> Any:
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, np.generic):
        return obj.item()
    elif isinstance(obj, dict):
        return {key: convert_np_types(value) for key, value in obj.items()}
    elif isinstance(obj, list):
        return [convert_np_types(item) for item in obj]
    else:
        return obj

# Initialize VideoWriter for output
def initialize_video_writer(source_video_path: str, output_video_path: str, fps: int = 30):
    cap = cv2.VideoCapture(source_video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    cap.release()
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4
    return cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Draw annotations on frame
def annotate_frame(frame, players, players_team_id, ball, ball_detections, keypoints):
    # Annotate players
    for i, (pos, conf, team_id) in enumerate(zip(players.xyxy, players.confidence, players_team_id)):
        x1, y1, x2, y2 = map(int, pos)
        team_color = (0, 255, 0) if team_id == 0 else (255, 0, 0)  # Green for team 0, Blue for team 1
        cv2.rectangle(frame, (x1, y1), (x2, y2), team_color, 2)
        cv2.putText(frame, f"ID: {i}, Team: {team_id}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, team_color, 1)

    # Annotate ball
    for pos, conf in zip(ball, ball_detections.confidence):
        x, y = map(int, pos)
        cv2.circle(frame, (x, y), 5, (0, 0, 255), -1)  # Red circle for ball
        cv2.putText(frame, f"Ball: {conf:.2f}", (x + 10, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

    # Optionally, add pitch keypoints annotations
    if hasattr(keypoints, "keypoints"):
        for kp in keypoints.keypoints:
            x, y = map(int, kp)
            cv2.circle(frame, (x, y), 5, (255, 255, 0), -1)  # Cyan for pitch keypoints

    return frame

# Callbacks for detection
def detect_players(frame, model, team_classifier) -> Dict[str, Any]:
    result = model(frame, imgsz=1280, verbose=False)[0]
    detections = sv.Detections.from_ultralytics(result)
    players = detections[detections.class_id == PLAYER_CLASS_ID]
    players_team_id = team_classifier.predict(get_crops(frame, players))
    return {"players": players, "players_team_id": players_team_id}

def detect_pitch(frame: np.ndarray, model) -> sv.KeyPoints:
    result = model(frame, verbose=False)[0]
    return sv.KeyPoints.from_ultralytics(result)

def detect_ball(frame: np.ndarray, slicer: sv.InferenceSlicer) -> Dict[str, Any]:
    ball_detections = slicer(frame).with_nms(threshold=0.5)
    ball_positions = ball_detections.get_anchors_coordinates(anchor=sv.Position.BOTTOM_CENTER)
    return {"ball_detections": ball_detections, "ball_positions": ball_positions}

# Main processing pipeline
def process_video_with_callbacks(
    source_video_path: str,
    device: str,
    player_detection_callback: Callable,
    pitch_detection_callback: Callable,
    ball_detection_callback: Callable
) -> Iterator[Dict[str, Any]]:
    # Initialize models
    player_detection_model = YOLO(PLAYER_DETECTION_MODEL_PATH).to(device=device)
    pitch_detection_model = YOLO(PITCH_DETECTION_MODEL_PATH).to(device=device)
    ball_detection_model = YOLO(BALL_DETECTION_MODEL_PATH).to(device=device)

    # Train the team classifier
    frame_generator = sv.get_video_frames_generator(source_path=source_video_path, stride=STRIDE)
    crops = []
    for frame in tqdm(frame_generator, desc="Collecting crops"):
        result = player_detection_model(frame, imgsz=1280, verbose=False)[0]
        detections = sv.Detections.from_ultralytics(result)
        crops += get_crops(frame, detections[detections.class_id == PLAYER_CLASS_ID])
    team_classifier = TeamClassifier(device=device)
    team_classifier.fit(crops)

    # Setup ball detection slicer
    def ball_callback(image_slice: np.ndarray) -> sv.Detections:
        result = ball_detection_model(image_slice, imgsz=640, verbose=False)[0]
        return sv.Detections.from_ultralytics(result)

    slicer = sv.InferenceSlicer(callback=ball_callback, slice_wh=(640, 640))

    # Process frames
    frame_generator = sv.get_video_frames_generator(source_path=source_video_path, stride=30)
    for frame_index, frame in enumerate(frame_generator):
        timestamp = f"{frame_index / 30:.2f}"  # Assuming 30 FPS video

        # Run callbacks
        pitch_data = pitch_detection_callback(frame, pitch_detection_model)
        player_data = player_detection_callback(frame, player_detection_model, team_classifier)
        ball_data = ball_detection_callback(frame, slicer)

        # Yield frame and JSON data
        yield {
            "frame": frame,
            "json": {
                "frame_index": frame_index,
                "timestamp": timestamp,
                "detections": {
                    "players": [
                        {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2), "team_id": int(team_id)}
                        for pos, conf, team_id in zip(
                            player_data["players"].xyxy,
                            player_data["players"].confidence,
                            player_data["players_team_id"]
                        )
                    ],
                    "ball": [
                        {"x": int(pos[0]), "y": int(pos[1]), "confidence": round(conf, 2)}
                        for pos, conf in zip(
                            ball_data["ball_positions"],
                            ball_data["ball_detections"].confidence
                        )
                    ]
                },
                "keypoints": pitch_data.to_dict() if hasattr(pitch_data, "to_dict") else {},
            }
        }

# Process video and save JSON + Annotated Video
source_video_path = "/content/08fd33_0.mp4"
output_video_path = "/content/annotated_output.mp4"
device = "cuda"  # or "cpu"

video_writer = initialize_video_writer(source_video_path, output_video_path)
all_frame_data = []

for result in process_video_with_callbacks(
    source_video_path,
    device,
    detect_players,
    detect_pitch,
    detect_ball
):
    frame = result["frame"]
    frame_data = result["json"]

    # Annotate frame
    annotated_frame = annotate_frame(
        frame,
        frame_data["detections"]["players"],
        frame_data["detections"]["ball"],
        frame_data["keypoints"]
    )

    # Save annotated frame to video
    video_writer.write(annotated_frame)

    # Collect JSON data
    all_frame_data.append(convert_np_types(frame_data))

# Finalize video and JSON
video_writer.release()
json_output_path = "output01.json"
with open(json_output_path, "w") as json_file:
    json.dump(all_frame_data, json_file, indent=4)

print("Processing complete. Annotated video and JSON output saved.")
```

# Re-Identification


Kalman Filters


https://paperswithcode.com/task/person-re-identification




Match the detection with the stored tracker using a combination of:
1. Spatial proximity.
2. Appearance features (ReID embeddings).
3. Historical trajectory patterns.



```python
from inference import get_model
import supervision as sv
import cv2
import numpy as np

# Load the object detection model
PLAYER_DETECTION_MODEL = get_model(model_id="football-players-detection-3zvbc/11", api_key="MV9lmTMpz1bESUNpgrLl")

# Video and detection settings
SOURCE_VIDEO_PATH = "/content/121364_0.mp4"
BALL_ID = 0
PLAYER_ID = 2
MOTION_THRESHOLD = 10000  # Threshold for motion detection
STRIDE = 5 # Initial stride value

# Annotators
ellipse_annotator = sv.EllipseAnnotator(
    color=sv.ColorPalette.from_hex(['#00BFFF']),
    thickness=2
)
label_annotator = sv.LabelAnnotator(
    color=sv.ColorPalette.from_hex(['#00BFFF']),
    text_color=sv.Color.from_hex('#000000'),
    text_position=sv.Position.BOTTOM_CENTER
)
triangle_annotator = sv.TriangleAnnotator(
    color=sv.Color.from_hex('#FFD700'),
    base=25,
    height=21,
    outline_thickness=1
)

# Tracker
tracker = sv.ByteTrack()
tracker.reset()

# Frame generator
frame_generator = sv.get_video_frames_generator(SOURCE_VIDEO_PATH, stride=STRIDE)

# Get video information
video_info = sv.VideoInfo.from_video_path(SOURCE_VIDEO_PATH)
target_video_path_original = "output_video_testing.mp4"

# Interpolation function
def interpolate_detections(prev_detections, current_detections, stride):
    """
    Interpolates bounding boxes between frames to account for skipped frames.
    Args:
        prev_detections: Previous frame detections.
        current_detections: Current frame detections.
        stride: Number of skipped frames.
    Returns:
        Interpolated detections with matching dimensions.
    """
    if prev_detections is None or len(prev_detections.xyxy) == 0:
        return current_detections

    if len(prev_detections.xyxy) != len(current_detections.xyxy):
        # If the number of detections doesn't match, return current detections as-is.
        print("Mismatch in detection count. Skipping interpolation.")

        return current_detections

    interpolated_boxes = []
    for prev_box, curr_box in zip(prev_detections.xyxy, current_detections.xyxy):
        interpolated_boxes.append([
            prev_box + (curr_box - prev_box) * (i / stride)
            for i in range(1, stride + 1)
        ])

    interpolated_boxes = np.array(interpolated_boxes).reshape(-1, 4)
    current_detections.xyxy = np.vstack([prev_detections.xyxy, interpolated_boxes])
    return current_detections

# Process video and write to file
previous_detections = None
previous_frame = None

count_frameProcees = 0

with sv.VideoSink(target_path=target_video_path_original, video_info=video_info) as sink:
    for frame in frame_generator:
        count_frameProcees += 1
        # Adjust stride based on motion detection
        if previous_frame is not None:
            motion = cv2.absdiff(previous_frame, frame)
            motion_intensity = np.sum(motion)
            if motion_intensity > MOTION_THRESHOLD:
                STRIDE = max(5, STRIDE - 1)  # Decrease stride on high motion
            else:
                STRIDE = min(15, STRIDE + 1)  # Increase stride on low motion

        previous_frame = frame

        # Run inference
        result = PLAYER_DETECTION_MODEL.infer(frame, confidence=0.3)[0]
        detections = sv.Detections.from_inference(result)

        # Handle interpolation
        if previous_detections is not None:
            detections = interpolate_detections(previous_detections, detections, STRIDE)

        # Tracker update and annotation logic
        detections = tracker.update_with_detections(detections)

        # Separate ball and player detections
        ball_detections = detections[detections.class_id == BALL_ID]
        ball_detections.xyxy = sv.pad_boxes(xyxy=ball_detections.xyxy, px=10)

        player_detections = detections[detections.class_id != BALL_ID]
        labels = [
            f"#{tracker_id}"
            for tracker_id in player_detections.tracker_id
        ]

        # Annotate frame
        annotated_frame = frame.copy()
        annotated_frame = ellipse_annotator.annotate(
            scene=annotated_frame,
            detections=player_detections
        )
        annotated_frame = label_annotator.annotate(
            scene=annotated_frame,
            detections=player_detections,
            labels=labels
        )
        annotated_frame = triangle_annotator.annotate(
            scene=annotated_frame,
            detections=ball_detections
        )

        # Plot and write frame
        # sv.plot_image(annotated_frame)
        sink.write_frame(annotated_frame)

        # Store current detections for the next iteration
        previous_detections = detections

print(count_frameProcees)
```
